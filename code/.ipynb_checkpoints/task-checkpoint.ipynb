```python
import pandas as pd
```

```python
event = pd.read_csv('../data/event_interview.csv', index_col=None, header=0)
fighter = pd.read_csv('../data/fighter_interview.csv', index_col=None, header=0)
outcome = pd.read_csv('../data/outcome_interview.csv', index_col=None, header=0)
```

```python
event.head()
```

```python
fighter.head()
```

```python
outcome.head()
```

```python
event_datframe = pd.DataFrame(
    event)
#event_datframe
fighter_datframe = pd.DataFrame(
    fighter)
#fighter_datframe
outcome_datframe = pd.DataFrame(
    outcome)
#outcome_datframe
```

```python

#mergeing the outcome and event dataframe to get the event name
result = pd.merge(outcome_datframe, event_datframe, on='id')

#getting only required columns
result1 = result[['fighter1_id','fighter2_id','name']]
```

```python
#renaming the column names to get the required column names
result1.rename(columns={'fighter1_id': 'id','fighter2_id': 'fighter2_id','name': 'event_name'}, inplace=True)

#at this place we have the event name and two players ids
```

```python
#merge the uper result1 and fighter dataframe to get the name and nationality for the first player.
result1=pd.merge(result1, fighter_datframe, on='id')

#geting only required columns
result1=result1[['fighter2_id','event_name','name','nationality']]
```

```python
#checking result for own satisfaction
result1.head()
```

```python
#rename the dataframe' columns to make it our required answer and ready for next merge.
result1.rename(columns={'fighter2_id': 'id','name': 'event_name', 'name':'fighter1_name','nationality':'fighter1_nationality'}, inplace=True)

#getting only required columns
result1= result1[['fighter1_name', 'id', 'fighter1_nationality', 'event_name']]
result1.head()

```

```python
#merging the result1 and fighter to get the other fighter details.
result1=pd.merge(result1, fighter_datframe, on='id')
result1.head()
```

```python
#getting only required columns
result1= result1[['fighter1_name', 'fighter1_nationality', 'event_name','name','nationality']]

#renaming the columns
result1.rename(columns={'fighter1_name': 'fighter1_name','fighter1_nationality': 'fighter1_nationality', 'event_name':'event_name','name':'fighter2_name','nationality':'fighter2_nationality'}, inplace=True)



```

```python
#re-arraining the columns as per our requirements
result1=result1[['fighter1_name','fighter2_name','fighter1_nationality','fighter2_nationality','event_name']]
result1
result1.to_csv('usama_task.csv')
```
